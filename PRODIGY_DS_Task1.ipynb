# Task 1: Visualisation of Population Data

```python
import pandas as pd

metadata_indicator= '/content/Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_45183.csv'
metadata_country = '/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_45183.csv'
population_data = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_45183.csv'

metadata_indicator_df = pd.read_csv(metadata_indicator)
metadata_country_df = pd.read_csv(metadata_country)
population_data_df = pd.read_csv(population_data)

metadata_indicator_df.head()
```

```python
import matplotlib.pyplot as plt

# Select data for the year 2022 and drop rows with missing values
population_2022 = population_data_df[['Country Name', '2022']].dropna()

# Sort the data by population size
population_2022 = population_2022.sort_values(by='2022', ascending=False).head(20)

# Create a bar chart
plt.figure(figsize=(12, 8))
plt.barh(population_2022['Country Name'], population_2022['2022'], color='skyblue')
plt.xlabel('Population in 2022')
plt.ylabel('Country')
plt.title('Top 20 Countries by Population in 2022')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest population at the top
plt.show()
```
